```python
import numpy as np
from statsmodels.stats.weightstats import ztest
```

```python
pop_mean  = 10000
sam_mean = 10800
sample_std = 2000
n = 50
```

```python
data = np.random.normal(sam_mean, sample_std, n)
```

```python
data
```

```python
z_score, p_value = ztest(data, value = pop_mean)
```
