```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
df['gen_cat'] = df['sex'].astype('category').cat.codes
```

```python
df
```

```python
fig = plt.figure(figsize=(10,20))
ax = fig.add_subplot(111,projection = '3d') # 1st row, 1st plot, 1st column
ax.scatter(df['total_bill'], df['tip'], df['size'], c = df['gen_cat'])
ax.set_xlabel('total_bill')
ax.set_ylabel('tip')
ax.set_zlabel('size')
plt.show()
```
