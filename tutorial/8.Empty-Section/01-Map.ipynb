```python
def fahrenheit(celsius):
    return (9/5)*celsius + 32
    
temps = [0, 22.5, 40, 100]

F_temps = map(fahrenheit, temps)

#Show
##list(F_temps)

list(map(lambda x: (9/5)*x + 32, temps))
```
